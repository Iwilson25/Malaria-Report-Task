### IPTP3 Coverage

### EST Incidence

```python
# Define the path to the EST Incidence CSV file
csv_path = "../Dataset/EST_INCIDENCE.csv"

# Read the data from the CSV file
df_est_incidence = pd.read_csv(csv_path, index_col=0)

df_est_incidence.head()
```

### 1ST LINE TREATED

```python
# Define the path to the 1st Line Treated CSV file
csv_path = "../Dataset/1STLINE_TREATED.csv"

# Read the data from the CSV file
df_1stline_treated = pd.read_csv(csv_path, index_col=0)

df_1stline_treated.head()
```

## Import Country Code

```python
# Define the path to the country code CSV file
csv_path = "../Dataset/COUNTRY_CODE.csv"

# Read the data from the CSV file
df_country_code = pd.read_csv(csv_path, index_col=0)

# Rename column name
df_renamed_cc = df_country_code.rename(columns={'Code': 'SpatialDimensionValueCode'})

df_renamed_cc.head()
```

## Merge Dataset to Country Code 

```python
#1stline_treated
# Merge the two datasets using a common column
merged_1stline_treated = pd.merge(df_1stline_treated, df_renamed_cc, on='SpatialDimensionValueCode')

# Save the merged dataset as a new CSV file
merged_1stline_treated.to_csv('../Dataset/merged_1stline_treated.csv', index=False)
```

```python
#EST Incidence
# Merge the two datasets using a common column
merged_est_incidence = pd.merge(df_est_incidence, df_renamed_cc, on='SpatialDimensionValueCode')

# Save the merged dataset as a new CSV file
merged_est_incidence.to_csv('../Dataset/merged_est_incidence.csv', index=False)

```

```python
#EST Mortality
# Merge the two datasets using a common column
merged_est_mortality = pd.merge(df_est_mortality, df_renamed_cc, on='SpatialDimensionValueCode')

# Save the merged dataset as a new CSV file
merged_est_mortality.to_csv('../Dataset/merged_est_mortality.csv', index=False)

```

```python
#IPTP3 Coverage
# Merge the two datasets using a common column
merged_IPTP3_coverage = pd.merge(df_iptp3_coverage, df_renamed_cc, on='SpatialDimensionValueCode')

# Save the merged dataset as a new CSV file
merged_IPTP3_coverage.to_csv('../Dataset/merged_IPTP3_coverage.csv', index=False)

```

```python
#ITN Coverage
# Merge the two datasets using a common column
merged_ITN_coverage = pd.merge(df_itn_coverage, df_renamed_cc, on='SpatialDimensionValueCode')

# Save the merged dataset as a new CSV file
merged_ITN_coverage.to_csv('../Dataset/merged_ITN_coverage.csv', index=False)

```

```python
#Total Cases
# Merge the two datasets using a common column
merged_total_cases = pd.merge(df_total_cases, df_renamed_cc, on='SpatialDimensionValueCode')

# Save the merged dataset as a new CSV file
merged_total_cases.to_csv('../Dataset/merged_total_cases.csv', index=False)

```
